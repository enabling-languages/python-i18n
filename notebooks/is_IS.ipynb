# Íslensk sveitarfélög eftir mannfjölda

Icelandic municipalities by population.

```python
import os, sys
import pandas as pd
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import unicodedata as ud
import regex as re

# are_equivalent(list1, list2) or are_equivalent(string1, string2)
#    compares two lists or two strings for cannonical equivalence.
#
def are_equivalent(i1, i2):
    if isinstance(i1, list) and isinstance(i2, list):
      return [ud.normalize("NFD", e) for e in i1] == [ud.normalize("NFD", e) for e in i2]
    elif isinstance(i1, str) and isinstance(i2, str):
      return ud.normalize("NFD", i1) == ud.normalize("NFD", i2)
    return

PAGE = "https://is.wikipedia.org/wiki/%C3%8Dslensk_sveitarf%C3%A9l%C3%B6g_eftir_mannfj%C3%B6lda"
caption = 'Íslensk sveitarfélög eftir mannfjölda'
table = pd.read_html(PAGE, match=caption, encoding="utf-8", thousands=" ", decimal=",", header=0)
```

```python
df = table[0]
def clean_labels(s):
    #s = re.sub(r'[\p{Pd}\p{Ps}\p{Pe}\p{Pi}\p{Pf}\p{Po}]+','',s) # Remove punction, except \p{Pc}
    s = re.sub(r'\(.+\)', '', s)
    s = re.sub(r'\[\d\]', '', s)
    #s = re.sub(r'[\p{P}±]+', '', s) # Strip punctuation
    s = re.sub(r'\p{Z}+', '_', s) # Replacea sequence of whitespace characters with a single underscore..
    # s = re.sub(r'^_|_$', '', s) # remove string initial or final underscores
    return s
df.rename(clean_labels, axis=1, inplace=True)

df_new = df.copy()
df_new.set_index('Nafn', inplace=True)
df_new.head(10)
```

## 2) Default sort

```python
df_sorted = df_new.sort_index()
sorted_list = df_sorted.index.tolist()
print(sorted_list)
#df_sorted.tail(10)
```

## Locale based sort

```python
import locale
loc = "is_IS.UTF-8"
locale.setlocale(locale.LC_COLLATE, loc)
#locale.setlocale(locale.LC_COLLATE, "en_AU.UTF-8")

df_sorted_locale = df_new.sort_index(key=lambda x: x.map(locale.strxfrm))

locale_list = df_sorted_locale.index.tolist()
print(are_equivalent("Equivantent? ", sorted_list, locale_list))
print(locale_list)
#df_sorted_locale.tail(10)
```

```python
from icu import Locale, Collator

loc = Locale("is_IS")
collator = Collator.createInstance(loc)

df_sorted_icu = df_new.sort_index(key=lambda x: x.map(collator.getSortKey))

icu_list = df_sorted_icu.index.tolist()
#df_sorted_icu.tail(10)

print("Equivantent? ", are_equivalent(sorted_list, icu_list))
print("Equivantent? ", are_equivalent(locale_list, icu_list))
```
